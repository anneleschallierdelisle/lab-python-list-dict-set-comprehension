````python
# calculate price for way
 
def calculate_price(customer_orders):
    list_prices = []
    for product in customer_orders:
        price = int(input(f'Enter the price of the ordered {product}: '))
        list_prices.append(price)
        total = sum(list_prices)
    return total

calculate_price(customer_orders)
````

````python
# calculate price comprehension
# customer_orders = ['mug', 'book']
def calculate_price(customer_orders):
    list_prices = {product: int(input(f"Enter the price of the ordered {product}")) for product in customer_orders}
    total = sum(list_prices.values())
    return total


total_price = calculate_price(customer_orders)
print(total_price)
````
